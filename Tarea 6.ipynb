La ecuacion a resolver es: $f(x,y) = e^{xy} + x^2y^2 + 5$         **cuando:** $(a = 0, b = 0, k = 2)$

La Serie de Taylor nos dice que:

$$f(x,y) = f(a,b) + \frac{\partial f(a,b)}{\partial x}(x-a) + \frac{\partial f(a,b)}{\partial y}(y-b) + \frac{1}{2!}[\frac{\partial^2 f(a,b)}{\partial x^2}(x-a)^2 + \frac{\partial^2 f(a,b)}{\partial y^2}(y-b)^2 + \frac{\partial^2 f(a,b)}{\partial x \partial y}(x-a)(y-b) + \frac{\partial^2 f(a,b)}{\partial y \partial x}(y-b)(x-a)]$$

Los pasos a seguir para resolverlo es:
1. $f(a,b) = e^{0\cdot 0} + 0^20^2 + 5 = 6$
2. $\frac{\partial f(a,b)}{\partial x} = 0\cdot e^{0\cdot 0} + 2\cdot 0\cdot 0^2 = 0$
3. $\frac{\partial f(a,b)}{\partial y} = 0\cdot e^{0\cdot 0} + 2\cdot 0\cdot 0^2 = 0$
4. $\frac{\partial^2 f(a,b)}{\partial x^2} = 0^2\cdot e^{0\cdot 0} + 2\cdot 0^2 = 0$
5. $\frac{\partial^2 f(a,b)}{\partial y^2} = 0^2\cdot e^{0\cdot 0} + 2\cdot 0^2 = 0$
6. $\frac{\partial^2 f(a,b)}{\partial x \partial y} = 0\cdot e^{0\cdot 0} + 4\cdot 0\cdot 0 = 1$
7. $\frac{\partial^2 f(a,b)}{\partial y \partial x} = 0\cdot e^{0\cdot 0} + 4\cdot 0\cdot 0 = 1$

Ya que tenemos todos los valores podemos sustituir en la Serie de Taylor.

$$f(x,y) = 6 + 0(x-0) + 0(y-0) + \frac{1}{2!}[0(x-0)^2 + 0(y-0)^2 + 0(x-0)(y-0) + 1(y-0)(x-0)] = 6+xy$$
